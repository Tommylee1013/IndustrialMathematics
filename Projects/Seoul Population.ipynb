```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
seoul = pd.read_csv('../Data/pop_ratio.csv', encoding = 'cp949')
```

```python
seoul = seoul[1:452]
```

```python
seoul
```

```python
def remove_parentheses(text):
    return text.split('(')[0]

seoul['행정구역'] = seoul['행정구역'].apply(remove_parentheses)
```

```python
seoul['행정구역'] = seoul['행정구역'].str.replace('서울특별시', '').str.replace(' ','')
```

```python
seoul.index = seoul.행정구역
seoul.drop('행정구역', axis = 1, inplace = True)
```

```python
for i in seoul.columns :
    try : seoul[i] = seoul[i].str.replace(',', '')
    except : continue
```

```python
seoul = seoul.astype(int)
```

```python
seoul['노인'] = seoul.iloc[:,67:103].sum(axis = 1)
```

```python
new_seoul = pd.DataFrame(seoul['노인']/seoul['2023년09월_계_총인구수'], columns = ['elder_ratio'])
```
